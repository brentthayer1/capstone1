```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import capstone1_funcs as cs1
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('/Volumes/b/Galvanize/DS-RFT4/capstones-RFT4/datasets/music2/SpotifyAudioFeaturesNov2018.csv')
df2 = pd.read_csv('/Volumes/b/Galvanize/DS-RFT4/capstones-RFT4/datasets/music2/SpotifyAudioFeaturesApril2019.csv')
```

```python
frames = [df1, df2]

df = pd.concat(frames)
```

```python
df.head()
```

Need to check for duplicate songs

```python
df.groupby()
```

```python
duplicates = df.groupby(['artist_name', 'track_name'], as_index=True).size()
duplicates[duplicates > 1].count()
```

```python
df.drop_duplicates(subset=['artist_name','track_name'], inplace=True)
```

```python
df.info()
```

```python
df['popularity'].isnull().sum()
```

```python
fig, ax_list = plt.subplots(13,1, figsize=(20,30))
cols = ['acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']
for ax, col in zip(ax_list, cols):
    df[col].hist(ax=ax, bins=125)
    ax.set_title(col)

fig.tight_layout()

```
