```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import capstone1_funcs as cs1
```

```python
df1 = pd.read_csv('/Volumes/b/Galvanize/DS-RFT4/capstones-RFT4/datasets/music2/SpotifyAudioFeaturesNov2018.csv')
df2 = pd.read_csv('/Volumes/b/Galvanize/DS-RFT4/capstones-RFT4/datasets/music2/SpotifyAudioFeaturesApril2019.csv')
```
